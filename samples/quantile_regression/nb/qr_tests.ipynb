# Normal distribution

```python
from scipy.stats import norm
```

```python
x = np.arange(-10, 10, 0.1)
```

```python
c = norm.cdf(x, loc=0, scale=1)
```

```python
plt.plot(x, c)
plt.title("CDF of normal distr");
```

```python
ic = norm.ppf(c)
```

```python
plt.plot(c, ic)
plt.title("Inverse CDF");
```

```python
d = c[1:] - c[:-1]
```

```python
plt.plot(d)
plt.title("Normal distribution");
```
